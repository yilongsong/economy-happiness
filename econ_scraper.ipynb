```python
#load packages
import pandas as pd
import re
import string
```

```python
#load the economic indicators file from IMF
df = pd.read_csv('WEO_Data.csv')
df.head()
```

```python
#standardise column names
def standardise_column_names(df, remove_punct=True):
    translator = str.maketrans(string.punctuation, ' '*len(string.punctuation))

    for c in df.columns:
        c_mod = c.lower()
        if remove_punct:            
            c_mod = c_mod.translate(translator)
        c_mod = '_'.join(c_mod.split(' '))
        if c_mod[-1] == '_':
            c_mod = c_mod[:-1]
        c_mod = re.sub(r'\_+', '_', c_mod)
        df.rename({c: c_mod}, inplace=True, axis=1)
    return df

```

```python
standardise_column_names(df)
```

```python
#deal with missing values
df["subject_descriptor"] = df["subject_descriptor"].fillna("")
#df
```

```python
#print(df["subject_descriptor"].unique())
```

```python
#filter by subject descriptor
mask1 = df["subject_descriptor"].str.contains("Gross domestic product per capita, current prices")
mask2 = df["subject_descriptor"].str.contains("Inflation, average consumer prices")
mask3 = df["subject_descriptor"].str.contains("Unemployment rate")
#combine all the rows that satisfy any of the masks
df = df[mask1 | mask2 | mask3]

#filter by units
mask4 =  df["units"].str.contains("U.S. dollars")
mask5 =  df["units"].str.contains("Index")

#combine all the rows that satisfy any of the masks
df = df[mask3 | mask4 | mask5]
df.drop(['scale', 'country_series_specific_notes'], axis = 1, inplace = True)
df
```

```python
#load unemployment rate file
unemp = pd.read_csv('unemployment.csv', skiprows=[0], header = 1)
standardise_column_names(unemp)
```

```python
# create a list of columns to drop
cols_to_drop = [str(year) for year in range(1960, 2010)]

# select string columns using boolean indexing and combine them with selected year columns
unemp_filtered = unemp.loc[:, ~unemp.columns.isin(cols_to_drop) | unemp.columns.str.contains('country_name')]

unemp_filtered.drop(['country_code', 'indicator_name','indicator_code', 'unnamed_66'], axis = 1, inplace = True)
unemp_filtered
```
